```python
cleaned_penguin_df = penguins.dropna()
cleaned_penguin_df
```

```python
cleaned_penguin_df.isna().sum()
```

```python
sns.scatterplot(data = cleaned_penguin_df, x="flipper_length_mm", y="body_mass_g",hue='island')
plt.title("Flipper Length vs Body Mass")
plt.show()
```

```python
sns.pairplot(data=cleaned_penguin_df,hue='species')
plt.suptitle("Pair Plot of Penguin Features", y=1.02)
plt.show()
```

```python
sns.histplot(data=cleaned_penguin_df, x="body_mass_g", hue="species", kde=True)
plt.title("Body Mass Distribution")
plt.show()
```

```python
sns.boxplot(data=cleaned_penguin_df, x="species", y="body_mass_g", hue="sex")
plt.title("Boxplot")
plt.show()
```

```python
sns.violinplot(data=cleaned_penguin_df, x="species", y="flipper_length_mm", hue="sex", split=True)
plt.title("Violin Plot")
plt.show()
```

```python
sns.barplot(data=cleaned_penguin_df, x="island", y="bill_length_mm")
plt.title("Average Bill Length per Island")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=cleaned_penguin_df, x="species", hue="sex")
plt.title("Countplot")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=cleaned_penguin_df, x="species", hue="island")
plt.title("Number of Penguins by Species and Island")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.stripplot(data=cleaned_penguin_df, x="species", y="body_mass_g", hue="sex")
plt.title("Body Mass by Species and Sex")
plt.show()
```
